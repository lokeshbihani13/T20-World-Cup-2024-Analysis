<h1 align="center"> T20 World Cup Cricket Data Pre Processing </h1>

```python
#import necessary libraries
import pandas as pd
import json
```

<h4 style="color:white">(1) Process Match Results</h4>

```python
with open('web-scraping-result-JSON/t20_wc_match_results_2024.json') as f:
    data = json.load(f)

df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head(10)
```

```python
df_match.info()
```

**Use scorecard as a match id to link with other tables**

```python
df_match.rename({'scorecard': 'match_id'}, axis = 1, inplace = True)
df_match.head()
```

**Create a match ids dictionary that maps team names to a unique match id. This will be useful later on to link with other tables**

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
```

```python
df_match.to_csv('T20_wc2024_dataset_cleaned_CSV/t20_wc_match_summary_2024.csv', index = False)
```

<h4 style="color:white">(2) Process Batting Summary</h4>

```python
with open('web-scraping-result-JSON/t20_wc_batting_summary_2024.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary'])
  
df_batting = pd.DataFrame(all_records)
df_batting.head(11)
```

```python
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: "not_out" if pd.isna(x) else "out")
```

```python
df_batting['dismissal'] = df_batting.dismissal.fillna("not out")
df_batting.head(11)
```

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
df_batting.head()
```

**Cleanup weird characters**

```python

df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('†', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('(c)', '')) #captain

df_batting.sample(10)
```

```python
df_batting.shape
```

```python
df_batting.to_csv('T20_wc2024_dataset_cleaned_CSV/t20_wc_bating_summary_2024.csv', index = False)
```

<h4 style="color:white">(3) Process Bowling Summary</h4>

```python
with open('web-scraping-result-JSON/t20_wc_bowling_summary_2024.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
all_records[:2]
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_ids_dict)
df_bowling.head()
```

```python
df_bowling.to_csv('T20_wc2024_dataset_cleaned_CSV/t20_wc_bowling_summary_2024.csv', index = False)
```

<h4 style="color:white">(4) Process Players Information</h4>

```python
with open('web-scraping-result-JSON/t20_wc_player_info_2024.json') as f:
    data = json.load(f)
```

```python
df_players = pd.DataFrame(data)

print(df_players.shape)
df_players.info()
```

**Cleanup weird characters**

```python
df_players['name'] = df_players['name'].apply(lambda x: x.replace('â€', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('†', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('\xa0', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('(c)', ''))

df_players.head(10)
```

```python
df_players.drop('input', axis = 1, inplace = True)

```

```python
df_players[df_players['team'] == 'India']
```

```python
df_players.to_csv('T20_wc2024_dataset_cleaned_CSV/t20_wc_player_info_2024.csv', index = False)
```
